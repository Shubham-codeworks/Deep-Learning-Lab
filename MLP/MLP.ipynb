```python
from sklearn.datasets import make_blobs
```

```python
import numpy as np
```

```python
X_toy, y_toy = make_blobs(n_samples=1000, n_features=2, centers=2, random_state=42)
```

```python
X_toy.shape
```

```python
X_toy
```

```python
y_toy.shape
```

```python
import matplotlib.pyplot as plt
plt.scatter(x=X_toy[:, 0], 
            y=X_toy[:, 1], 
            c=y_toy, 
            cmap=plt.cm.RdYlBu);
plt.show()
```

```python
from sklearn.datasets import make_circles
```
